```python
import gymnasium as gym

env = gym.make("CarRacing-v3", 
               render_mode="rgb_array", 
               lap_complete_percent=0.95, 
               domain_randomize=False, 
               continuous=False)

obs, _ = env.reset()

randomize_obs, _ = env.reset(options={"randomize": True})
```

```python
import matplotlib.pyplot as plt

print(f"Observation shape: {obs.shape}, dtype: {obs.dtype}")  # Should print (96, 96, 3), uint8

# Get randomized observation
randomize_obs, _ = env.reset(options={"randomize": True})

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot regular observation - make sure to show the full 96x96 image
axes[0].imshow(obs)
axes[0].set_title('Regular Observation')
axes[0].axis('off')

# Plot randomized observation
axes[1].imshow(randomize_obs)
axes[1].set_title('Randomized Observation')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
